# Data Cleaning 

```python
def FixSeason(text):
    index_of_hyphen = text.find("-")
    
    if index_of_hyphen != -1:
        text = text[:index_of_hyphen]
        
    return int (text)
```

```python
df['Season'] = df['Season'].apply(FixSeason)
```

```python
df
```

# fixing Full Time Result (FTR) && Half Time Result (HTR)
making Home win = 0  , Draw = 1  , Away win = 2


```python
def Full_HalfTimeResult(a):
    try:
        if a == 'H':
            return int (0)
        
        elif a == 'D':
            return int (1)

        elif a =='A':
             return int (2)

        else:
            return np.nan
    except:
            return np.nan
```

```python
df['HTR'] = df['HTR'].apply(Full_HalfTimeResult)
```

```python
df['FTR'] = df['FTR'].apply(Full_HalfTimeResult)
```

# Questions

# 1) which team has been the champion of the season ?

```python
def get_champion(season_data):
    home_wins = season_data[season_data['FTR'] == 0]['HomeTeam'].value_counts()
    away_wins = season_data[season_data['FTR'] == 2]['AwayTeam'].value_counts()
    total_wins = home_wins.add(away_wins, fill_value=0)
    champion = total_wins.idxmax()
    return pd.Series({'Champion': champion})

champions_by_season = df.groupby('Season').apply(get_champion, include_groups=False)
print(champions_by_season)
```

# 2) which team has been the most champion of all the seasons ? 

```python
def get_champion(season_data):
    home_wins = season_data[season_data['FTR'] == 0]['HomeTeam'].value_counts()
    away_wins = season_data[season_data['FTR'] == 2]['AwayTeam'].value_counts()
    total_wins = home_wins.add(away_wins, fill_value=0)
    champion = total_wins.idxmax()
    return pd.Series({'Champion': champion})

champions_by_season = df.groupby('Season').apply(get_champion, include_groups=False)

team_season_counts = champions_by_season['Champion'].value_counts()
most_successful_team = team_season_counts.idxmax()

seasons_of_most_successful_team = champions_by_season[champions_by_season['Champion'] == most_successful_team].index.tolist()

print("The team that has been the champion of the season the most number of times is:", most_successful_team)
print("Seasons in which the most successful team was the champion:", seasons_of_most_successful_team)
```

# 3 ) which season have the highest number of goals ?

```python
goals_by_season = df.groupby('Season')[['FTHG', 'FTAG']].sum()

goals_by_season['TotalGoals'] = goals_by_season['FTHG'] + goals_by_season['FTAG']

sns.set(rc={'figure.figsize':[10,10]}, font_scale=1.2)
sns.barplot(x='TotalGoals', y=goals_by_season.index, data=goals_by_season, orient='h')
plt.xticks(rotation=90) 
plt.xlabel('Season')
plt.ylabel('Total Goals')
plt.title('Total Goals by Season')
plt.show()
```

# 4 ) Which team has more wins in away match ? 

```python
away_wins = df[df['FTR'] == 2]

away_wins_count = away_wins['AwayTeam'].value_counts()

sns.set(rc={'figure.figsize':[10,10]},font_scale=1.2)
sns.barplot(x=away_wins_count.values, y=away_wins_count.index)
plt.xticks(rotation=90)  
plt.xlabel('Away Team')
plt.ylabel('Number of Away Wins')
plt.title('Number of Away Wins by Team')
plt.show()
```

# 5 ) How many matches had no red cards shown to either team in each season?

```python
no_red_cards_matches = df[(df['HR'] == 0) & (df['AR'] == 0)].groupby('Season').size().reset_index(name='NoRedCardsMatches')

plt.figure(figsize=(10, 8))
sns.heatmap(data=no_red_cards_matches.pivot(index='Season', columns='NoRedCardsMatches', values='NoRedCardsMatches'), annot=True, fmt='0.0f', cmap='YlGnBu')
plt.xlabel('Number of Matches with No Red Cards')
plt.ylabel('Season')
plt.title('Number of Matches with No Red Cards Shown in Each Season')
plt.show()
```

# 6) Which team has the highest number of red cards in all seasons ?

```python

home_red_cards = df.groupby('HomeTeam')['HR'].sum()
away_red_cards = df.groupby('AwayTeam')['AR'].sum()

total_red_cards = home_red_cards.add(away_red_cards, fill_value=0)

team_with_highest_red_cards = total_red_cards.idxmax()
highest_red_cards_count = total_red_cards.max()

print("Team with the highest number of red cards:", team_with_highest_red_cards)
print("Number of red cards:", highest_red_cards_count)
```

# 11) highest number of goals of each team in one season ?


```python
import pandas as pd

team_goals = pd.concat([df.groupby(['Season', 'HomeTeam'])['FTHG'].sum(),
                        df.groupby(['Season', 'AwayTeam'])['FTAG'].sum()])

max_goals_by_team_in_season = team_goals.groupby(level=1).max().reset_index()
max_goals_by_team_in_season.columns = ['Team', 'MaxGoals']

plt.figure(figsize=(12, 8))
sns.set(rc={'figure.figsize':(12,8)}, font_scale=1.2)
sns.barplot(x='MaxGoals', y='Team', data=max_goals_by_team_in_season, hue='Team', dodge=False, legend=False)

plt.xlabel('Maximum Goals in a Season')
plt.ylabel('Team')
plt.title('Maximum Goals Scored by Each Team in a Single Season')
plt.show()
```

# 16) How many fouls committed by each team

```python
all_teams = pd.concat([df['HomeTeam'], df['AwayTeam']])

# Concatenate fouls committed by home team and away team for each match
all_fouls = pd.concat([df['HF'], df['AF']])

# Create a DataFrame to store the total fouls committed by each team
fouls_by_team = pd.DataFrame({'Team': all_teams, 'Fouls': all_fouls})

# Group by team and sum the fouls
total_fouls_by_team = fouls_by_team.groupby('Team')['Fouls'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Fouls', y='Team', data=total_fouls_by_team, hue='Team', dodge=False, legend=False)

# Set labels and title
plt.xlabel('Total Fouls Committed')
plt.ylabel('Team')
plt.title('Total Fouls Committed by Each Team')

# Show plot
plt.show()
```

# 17) What is the 5 Most Common Results Based on Total Goals

```python
total_goals = df['FTAG'] + df['FTHG']

# Count the occurrences of each total goals value
common_results = total_goals.value_counts().head(5)

plt.figure(figsize=(10, 6))
sns.countplot(x=total_goals, order=common_results.index, hue=total_goals, dodge=False, legend=False, palette='viridis')

# Set labels and title
plt.xlabel('Total Goals')
plt.ylabel('Frequency')
plt.title('Top 5 Most Common Results Based on Total Goals')

# Show plot
plt.show()
```
